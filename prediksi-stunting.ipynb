```python
dataset_zip = zipfile.ZipFile('stunting.zip','r')
dataset_zip.extractall()
dataset_zip.close()
```

```python
dataset = pd.read_excel('Stunting Dataset (1).xlsx')
```

```python
dataset
```

```python
x_train, y_train = dataset.loc[:, 'Sex':'ASI Eksklusif'], dataset['Stunting']
x_train['Sex'] = x_train['Sex'].replace({'F':0, 'M':1})
x_train['ASI Eksklusif'] = x_train['ASI Eksklusif'].replace({'No':0, 'Yes':1})
```

```python
y_train = y_train.replace({'No': 0, 'Yes': 1})
```

```python
print({'x_train': x_train})
print({'y_train': y_train})
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, shuffle=True)
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation='relu', input_shape=(7,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

model.summary()

history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=500)
```

```python
_, accuracy_training = model.evaluate(x_train, y_train)
_, accuracy_testing = model.evaluate(x_test, y_test)

print('Accuracy Training:', accuracy_training)
print('Accuracy Testing:', accuracy_testing)
```

```python
history = history.history

train_acc = history['accuracy']
train_loss = history['loss']
test_acc = history['val_accuracy']
test_loss = history['val_loss']

epochs_train_acc = range(len(train_acc))
epochs_train_loss = range(len(train_loss))
epochs_test_acc = range(len(test_acc))
epochs_test_loss = range(len(test_loss))

def plot_grafik(epochs, data, name):
  plt.plot(epochs, data, 'r')
  plt.title(name)
  plt.xlabel('Epochs')
  plt.ylabel(name)
  plt.legend([name])
  plt.figure()

plot_grafik(epochs_train_acc, train_acc, 'Training Accuracy')
plot_grafik(epochs_train_loss, train_loss, 'Training Loss')
plot_grafik(epochs_test_acc, test_acc, 'Testing Accuracy')
plot_grafik(epochs_test_loss, test_loss, 'Testing Loss')
```
